```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c = '#a366ff', alpha=.9,marker = 'X')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10 ,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2)) # change the size of figure
plt.plot(x,y, '--r') # dash line graph
plt.xlabel("this is my x data") # give label to x axis
plt.ylabel("this is my y data")  # give label to y axis
plt.title("this is my sin wave graph") # give title to graph
plt.grid() # show grap in form of grid
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
# categorical vs numerical data using bar graph
plt.figure(figsize=(5,3)) # sabse phele fig size ko likhte hai
plt.bar(x,y,color='#cc00cc')
plt.xlabel("Representing my categorical data")
plt.ylabel("Representing my numerical data")
plt.title("This is bar plot")
plt.grid()
```

```python
# categorical vs numerical data using bar graph horizontal use barh
plt.figure(figsize=(5,3)) # sabse phele fig size ko likhte hai
plt.barh(x,y,color='#cc00cc')
plt.ylabel("Representing my categorical data")
plt.xlabel("Representing my numerical data")
plt.title("This is bar plot")
plt.grid()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c ='#9327E8', alpha = .9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
